```python
import pygame
import sys
import random
import time

class snake():
    def __init__(self):
        self.position = [100,50]
        self.body = [[100,50],[90,50],[80,50]]
        self.direction = "RIGHT"
        self.changeDirectionto = self.direction
    
    def changeDir(self,dir):
        if dir=='RIGHT' and not self.direction == 'LEFT':
            self.direction = 'RIGHT'
        if dir=='LEFT' and not self.direction == 'RIGHT':
            self.direction = 'LEFT'
        if dir=='UP' and not self.direction == 'DOWN':
            self.direction = 'UP'
        if dir=='DOWN' and not self.direction == 'UP':
            self.direction = 'DOWN'

    def move(self,foodpos):
        if self.direction == 'RIGHT':
            self.position[0] = self.position + 10
        if self.direction == 'LEFT':
            self.position[0] = self.position - 10
        if self.direction == 'DOWN':
            self.position[1] = self.position - 10
        if self.direction == 'UP':
            self.position[1] = self.position + 10
    
        ###increase one block with movement and check if next block is food or not
        self.body.insert(0,list(self.position))
        
        if self.position==foodpos:
            return 1
        else:
            self.body.pop()
            return 0
        
    def checkcollision(self):
        if(self.position[0]>490 or self.position[0]<10 or self.position[1]>490 or self.position[1]<10):
            return 1
        ###dont iterate for head
        for bodypart in self.body[1:]:
            if bodypart == self.position:
                return 1
        return 0
    
    def getheadpos(self):
        return self.position
    
    def getbody(self):
        return self.body
    
class foodmaker():
    def __init__(self):
        self.position = [random.randrange(1,50)*10,random.randrange(1,50)*10]
        self.foodexists = True
        
    def make_food(self):
        if self.foodexists == False:
            self.position = [random.randrange(1,50)*10,random.randrange(1,50)*10]
            self.foodexists = True
        return self.position
    
    def setfood(self,status):
        self.foodexists = status
        
    
    
window = pygame.display.set_mode((500,500))
fps = pygame.time.Clock()

score = 0

sn = snake()
foodm = foodmaker()

def game_over():
    pygame.quit()
    sys.exit()
    
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            game_over()
        
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_DOWN:
                sn.changeDir('DOWN')
            if event.key == pygame.K_UP:
                sn.changeDir('UP')
            if event.key == pygame.K_LEFT:
                sn.changeDir('LEFT')
            if event.key == pygame.K_RIGHT:
                sn.changeDir('RIGHT')
                
            foodpos = foodm.make_food()
            if(sn.move(foodpos)==1):
                score = score+1
                foodm.setfood(False)
            
            window.fill((255,255,255))
            for pos in sn.getbody():
                pygame.draw.rect(window,(0,255,0),(pos[0],pos[1],10,10))
            pygame.draw.rect(window,(0,0,0),(foodpos[0],foodpos[1],10,10))
            
            if sn.checkcollision()==1:
                game_over()
            
            pygame.display.set_caption('Score: '+score)
            pygame.display.flip()
            fps.tick(24)
```
